```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
from matplotlib import font_manager, rc
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df = pd.read_csv('titanic/train.csv')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18,8))
ax[0].set_title('Survived Barplot')
ax[1].set_title('Survived Pieplot')

sns.countplot(data=df, x='Survived', ax=ax[0], palette='deep')
#Pallete kinds: deep, muted, pastel, bright, dark, colorblind, Set1~3, Paired, viridis, cubehelix, rocket, icefire, magma, inferno, plasma, cividis
ax[1].pie(df['Survived'].value_counts(), labels=['Dead','Survived'], autopct='%1.1f%%')
plt.show()
```
